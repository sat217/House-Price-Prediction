# House Price Prediction - Exploratory Data Analysis (EDA)
Basic EDA for House Prices dataset.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load Data
train_path = os.path.join("..", "data", "train.csv")
df = pd.read_csv(train_path)
```

```
# 1. Dataset Shape
print(f"Dataset Shape: {df.shape}")
```

```
# 2. Check Missing Values
missing = df.isnull().sum()
print("Missing values per column:\n", missing[missing > 0].sort_values(ascending=False))
```

```
# 3. Summary Statistics
display(df.describe())
```

```
# 4. Correlation Heatmap
plt.figure(figsize=(12, 10))
numeric_cols = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_cols.corr()
# We can just show top 15 correlated with SalePrice to keep it clean
cols = corr.nlargest(15, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, 
                 yticklabels=cols.values, xticklabels=cols.values)
plt.title("Correlation Heatmap")
plt.show()
```

```
# 5. Price Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['SalePrice'], bins=30, kde=True, color='blue')
plt.title('SalePrice Distribution')
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.show()
```
